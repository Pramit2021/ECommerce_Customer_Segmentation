1. IMPORT THE PANDAS LIBRARY

```python
import pandas as pd
```

2. READ THE CSV FILE

```python
data = pd.read_csv('Input/data.csv', encoding="ISO-8859-1")
data.head()
```

3. CLEANING THE DATA

```python
# Dataframe dimensions
data.shape
```

```python
# Looking at the data types
data.info()
```

```python
# looking for null values
data.isnull().sum()
```

```python
# Drop rows with missing CustomerID or Description
data_cleaned = data.dropna(subset=['CustomerID', 'Description'])
```

```python
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])
```

```python
data_cleaned.info()
```

```python
data_cleaned.head()
```

```python
data_cleaned.shape
```

```python
# worksheet for tableau dashboard
data_cleaned.to_csv('Customer_segmentation_database.xlsx')
```

4. DATAFRAME EXPORT

```python
y = []

for i in range(len(data_cleaned)):
    x = tuple(data_cleaned.iloc[i])
    y.append(x)

y
```

5. SAVE THE FILE AS TEXT FILE

```python
file = open('Customer_Segmentation_data.txt', 'w')

for tuple in y:
    file.write(str(tuple) + ',' + '\n')

file.close()
```
